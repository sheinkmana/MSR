```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
class MeanSquaresRegression:
    def __init__(self, degree=3):
        self.w = 0
        self.deg = degree
    
    def train(self, x, y):
        X = np.array([x**k for k in range(self.deg+1)]).T
        self.w = np.linalg.inv(X.T@X)@X.T@y
        
    def predict(self, x):
        X = np.array([x**k for k in range(self.deg+1)]).T
        return X@self.w

another_model = MeanSquaresRegression(degree=2)
another_model.train(x,y)

dummy_model = MeanSquaresRegression(degree=5)
dummy_model.w


dummy_model.train(x,y)
dummy_model.w

dummy_model.predict(x_range)

plt.figure(figsize=(10,8))
plt.scatter(x_range, f(x_range), label = "Truth")
plt.scatter(x, y, label = "Noisy samples from truth")
plt.scatter(x_range, dummy_model.predict(x_range), label = "Predictions")
plt.scatter(x_range, another_model.predict(x_range), label = "Dummy Predictions")
plt.legend()
plt.show()
```

```python
def mean_squares_regression(X, y):
    return np.linalg.inv(X.T@X)@X.T@y

def f_hat(x,w,deg=3):
    X = np.array([x**k for k in range(deg+1)]).T
    return X@w

def f(x):
    return np.sin(x)*np.power(x,2)


N = 50
eta = 0

deg = 3

# Data generation
x = 10*np.random.rand(N)
y = f(x) + eta*np.random.randn(N)

X = np.array([x**k for k in range(deg+1)]).T

w = mean_squares_regression(X, y)

x_range = np.linspace(0,10)

plt.figure(figsize=(10,8))
plt.scatter(x_range, f(x_range), label = "Truth")
plt.scatter(x, y, label = "Noisy samples from truth")
plt.scatter(x_range, f_hat(x_range, w, deg), label = "Predictions")
plt.legend()
plt.show()
```

```python
class ChildClass(MeanSquaresRegression):
    def __init__(self, noise=0):        
        super().__init__()
        self.noise = noise  

    
    def noise_predict(self, x):
        return super().predict(x) + self.noise*np.random.randn()


model = ChildClass(noise=2)

model.train(x,y)

plt.figure(figsize=(10,8))
plt.scatter(x_range, f(x_range), label = "Truth")
plt.scatter(x, y, label = "Noisy samples from truth")
plt.scatter(x_range, model.noise_predict(x_range), label = "Inherited Predictions")
plt.scatter(x_range, dummy_model.predict(x_range), label = "Predictions")
plt.legend()
plt.show()
```
